# Support Vector Machine

A Support Vector Machine (SVMs) are one of the most popular supervised learning models with associated learning algorithms that analyze data and recognize patterns, used for classification and regresison analysis.

In the SVM algorithm, we plot each data item as a point in n-dimensional space (where n is a number of features you have) with the value of each feature being the value of a particular coordinate. Then, we perform classification by finding the hyper-plane that differentiates the two classes very well.

SVM chooses the extreme points/vectors that help in creating the hyperplane. These extereme cases are called as support vecotrs, and hence algoirthm is termed as Support Vector Machines. 

### Imputing Missing Data with IterativeImputer and using SVM on the unscaled data

```python
fig = plt.figure(figsize=(15,25))
sns.pairplot(df,hue='diabetes')
```

The above pairplot shows that the data is not linearly separable.

Separating into independent (X) and dependenty (y) variables.

```python
X = df.drop(['diabetes'],axis=1)
y= df['diabetes']
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#missing_values=all occurences of missing_values will be imputed.
#n_nearest_features is set to 8 so that all 8 features are used to estimate
#the missing values of each feature column.
#imputation_order is set to ascending so that the order in which the
#features  will be imputed will be form features with fewest
#missing values to most
imp = IterativeImputer(missing_values=np.nan,n_nearest_features=8,
                   initial_strategy='mean',
                   imputation_order='ascending',verbose=1)
svm = SVC()
steps = [
    ('iterative_imputer',imp),
    ('svm',svm)
]
pipeline = Pipeline(steps)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,
                                                random_state=42,
                                                stratify=y)
pipeline.fit(X_train,y_train)
```

```python
predictions = pipeline.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(f'Accuracy Score: {accuracy_score(y_test,predictions)}')
```

### Imputing Missing Data with IterativeImputer and using SVM on the scaled data

```python
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler,MinMaxScaler

#missing_values=all occurences of missing_values will be imputed.
#n_nearest_features is set to 8 so that all 8 features are used to estimate
#the missing values of each feature column.
#imputation_order is set to ascending so that the order in which the
#features  will be imputed will be form features with fewest
#missing values to most
scaler = MinMaxScaler()
imp = IterativeImputer(missing_values=np.nan,n_nearest_features=8,
                   initial_strategy='mean',
                   imputation_order='ascending',verbose=1)
svm = SVC(kernel='linear',probability=True)
steps = [
    ('iterative_imputer',imp),
    ('scaler',scaler),
    ('svm',svm)
]
pipeline = Pipeline(steps)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,
                                                random_state=42,
                                                stratify=y)
pipeline.fit(X_train,y_train)
```

```python
predictions = pipeline.predict(X_test)
predictions
```

```python
print(f'Accuracy Score: {accuracy_score(y_test,predictions)}')
```

```python
print(classification_report(y_test,predictions))
```

```python
predictions_prob = pipeline.predict_proba(X_test)[:,1]
predictions_prob
```

```python
from sklearn.metrics import roc_auc_score
```

```python
print(f'ROC-AUC score: {roc_auc_score(y_test,predictions_prob)}')
```

The roc_auc_score gives us the area under the Receiver Operating Characteristic Curve. The higher the area under the curver, the better is the model.

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, predictions_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr, tpr, label='Logistric Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.show()
```

### Using GridSearch to search for the best parameters for the Support Vector Machine

```python
from sklearn.model_selection import GridSearchCV
```

```python
steps = [
    ('iterative_imputer',IterativeImputer(missing_values=np.nan)),
    ('scaler',StandardScaler()),
    ('svm',SVC())
]

pipeline = Pipeline(steps)

parameters = {
    'svm__C': [0.1,1,10,100,1000], ## the strength of regularization parameter
    # is inversely propotinal to C.
    'svm__gamma': [1,0.1,0.01,0.001],
    'svm__kernel': ['poly','rbf',],
    'svm__probability': [True,False]
}

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=101,stratify=y)
cv = GridSearchCV(pipeline, param_grid=parameters, cv = 5)
cv.fit(X_train,y_train)

```

```python
cv
```

```python
cv.best_score_
```

```python
cv.best_params_
```

```python
predictions = cv.predict(X_test)
predictions
```

```python
print(f'Accuracy: {accuracy_score(y_test, predictions)}')
```

```python
print(classification_report(y_test, predictions))
```
