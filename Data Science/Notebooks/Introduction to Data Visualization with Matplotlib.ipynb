## Using Twin Axes

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change['relative_temp'])
ax2.set_ylabel('Relative Temperate (Celsius)')
plt.show()
```

### Separating Variabls by color

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change['relative_temp'], color='red')
ax2.set_ylabel('Relative Temp (Celsius)', color='red')
plt.show()
```

### Coloring the Ticks

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change['relative_temp'], color='red')
ax2.set_ylabel('Relative Temp (Celsius)', color='red')
ax2.tick_params('y', colors='red')
plt.show()
```

### Adding a legend

### Adding Error Bars to Bar Charts

```python
fig, ax = plt.subplots()
ax.bar('Rowing', 
       mens_rowing['Height'].mean(),
       yerr = mens_rowing['Height'].std()
      )
ax.bar('Gymnastics', 
       mens_gymnastics['Height'].mean(),
       yerr = mens_gymnastics['Height'].std()
      )
ax.set_ylabel('Height (cm)')
plt.show()
```

### Automating Figures from data

```python
summer = pd.read_csv('D:\Babin\Internship\Data Science\Datasets\summer2016.csv', index_col = 0)
summer.head()
```

```python
sports = summer['Sport'].unique()
sports
```

### Bar-chart of heights for all sports

```python
fig, ax = plt.subplots()
for sport in sports:
    sport_df  = summer[summer['Sport']==sport]
    ax.bar(sport, sport_df['Height'].mean(),
          yerr=sport_df['Height'].std())
ax.set_ylabel('Height (cm)')
ax.set_xticklabels(sport, rotation = 90)
plt.show()
```
