```python
tips = sns.load_dataset('tips')
tips.head()
```

### A Basic Scatter Plot

```python
sns.scatterplot(x='total_bill', 
               y='tip', 
               data=tips)
```

### A Scatter Plot with Hue

```python
sns.scatterplot(x='total_bill', 
               y='tip', 
               data=tips, 
               hue='smoker')
```

#### Settings Hue Order

```python
sns.scatterplot(
    x='total_bill', 
    y='tip', 
    data=tips,
    hue='smoker',
    hue_order=['No', 'Yes']
)
```

#### Specifying Hue Colors

```python
hue_colors = {
    'Yes': 'black',
    'No': 'red'
}
sns.scatterplot(
    x='total_bill',
    y='tip', 
    data = tips,
    hue='smoker',
    hue_order = ['No','Yes'],
    palette = hue_colors
)
```

#### Using HTML hex color codes with hue

```python
hue_colors = {"Yes": "#808080",
"No": "#00FF00"}
```

```python
sns.scatterplot(
    x='total_bill', 
    y='tip', 
    data = tips,
    hue='smoker', 
    hue_order = ['No', 'Yes'],
    palette = hue_colors
)
```

```python
masculanity_data = pd.read_csv('D:\Babin\Internship\Data Science\Datasets\masculinity.csv')
masculanity_data.head()
```

#### Changing the Order

```python
category_order = [
    'no answer', 
    'not at all', 
    'not very', 
    'Somewhat', 
    'very'
]

sns.catplot(
    x='how_masculine', 
    data = masculanity_data,
    kind='count',
    order = category_order
)
```

### Bar Plots

```python
sns.catplot(
    x='day', 
    y = 'total_bill', 
    data = tips, 
    kind='bar'
)
```

<ul>
    <li>Here, the line shows 95% confidence intervals for the mean.</li>
    <li>Shows uncertainty about our estimate</li>
    <li>Assumes our data is a random sample</li>
</ul>

### Turning off confidence intervals

```python
sns.catplot(
    x='day', 
    y = 'total_bill', 
    data = tips, 
    kind='bar',
    ci = None
)
```

### Changing the orientation

```python
sns.catplot(
    x='total_bill', 
    y = 'day', 
    data = tips, 
    kind='bar'
)
```

### Box Plot

```python
sns.catplot(
    x='time',
    y='total_bill',
    data = tips,
    kind='box'
)
```

### Change the order of categories

```python
sns.catplot(
    x='time', 
    y='total_bill', 
    data = tips, 
    kind='box',
)
```

```python
sns.catplot(
    x='time', 
    y='total_bill', 
    data = tips, 
    kind='box',
    order = ['Dinner', 'Lunch']
)
```

### Omitting the outliers using 'sym'

```python
sns.catplot(
    x='time', 
    y='total_bill', 
    data = tips, 
    kind='box',
    sym=''
)
```

### Changing the whiskers usingn 'whis'

<ul>
    <li>By default, the whiskers extend to 1.5 * interquartile range</li>
    <li>Make them extend to 2.0 * IQR: whis=2.0</li>
    <li>Show the 5th and 95th percentiles: whis=[5,95]</li>
    <li>Show the min and max values: whis=[0,100]</li>
</ul>

```python
tips[tips['time']=='Lunch']['total_bill'].min()
```

```python
tips[tips['time']=='Lunch']['total_bill'].max()
```

```python
tips[tips['time']=='Dinner']['total_bill'].min()
```

```python
tips[tips['time']=='Dinner']['total_bill'].max()
```

```python
sns.catplot(
    x='time', 
    y='total_bill', 
    data = tips, 
    kind='box', 
    whis=[0,100]a
)
```

### Point Plots

#### Point Plots vs Line Plots

<h3>Both Show:</h3>
<ul>
    <li>Mean of quantitative variable</li>
    <li>95% confidence intervals for the mean</li>
</ul>
<h3>Differences:</h3>
<ul>
    <li>Line plot has <b>quantitative variable</b> on x-axis.</li>
    <li>Point plot has <b>categorical</b> variable on x-axis</li>
</ul>

```python
tips.head()
```

```python
sns.catplot(
    x='time', 
    y='tip', 
    data = tips, 
    hue='smoker', 
    kind='point'
)
```
