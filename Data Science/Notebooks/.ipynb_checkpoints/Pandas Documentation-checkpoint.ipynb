# Using apply to change the duration of all the "Drama"

```python
copy_df = df.copy(deep=True)
copy_df.head()
```

```python
copy_df.dtypes
```

```python
copy_df.apply(lambda x: '50 minutes' if x.Genre=='Drama' else x.Duration, axis=1)
```
