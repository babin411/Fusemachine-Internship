### Defining a Color For a plot

<b>Note: </b> Seaborn's Grid Plots requires data in <b>Tidy</b> format.

```python
tuition = pd.read_csv('D:\Babin\Internship\Data Science\Datasets\\tuition.csv')
tuition.head()
```

```python
g = sns.FacetGrid(tuition, col='HIGHDEG')
g.map(sns.boxplot, 'Tuition', order=['1','2','3','4'])
```

```python
sns.factorplot(x='Tuition', data = tuition, col='HIGHDEG', kind='box')
```

### FacetGrid for regression

```python
tuition['HIGHDEG'].value_counts()
```

```python
g = sns.FacetGrid(tuition, col='HIGHDEG')
g.map(plt.scatter, 'Tuition', 'SAT_AVG_ALL')
```

### lmplot

```python
sns.lmplot(data= tuition, 
          x='Tuition', 
          y='SAT_AVG_ALL', 
          col='HIGHDEG', 
          fit_reg=False)
```

```python
sns.lmplot(data= tuition, 
          x='Tuition', 
          y='SAT_AVG_ALL', 
          col='HIGHDEG', 
          fit_reg=True)
```
