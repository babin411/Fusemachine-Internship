# Tensors

A tensor is basically a multi-dimensional array. 

## Initilialization of Tensors

```python
x = tf.constant(4)
x
```

```python
y = tf.constant(5.0)
y
```

```python
z = tf.constant(4, shape=(1,1))
z
```

```python
a = tf.constant(5, shape=(1,1), dtype=tf.float32)
a
```

```python
x = tf.constant([[1,2,3],[4,5,6]])
print(x)
```

```python
x = tf.ones((3,3))
x
```

```python
x = tf.zeros((2,3))
x
```

```python
x = tf.eye(3)
x
```

```python
x = tf.random.normal((3,3), mean=0, stddev=1)
x
```

```python
x = tf.random.uniform((1,3),minval=0, maxval=1)
x
```

```python
x = tf.range(9)
x
```

```python
x = tf.range(start=1, limit=10, delta=2)
x
```

## Casting to different types

```python
x = tf.cast(x, dtype=tf.float64)
x
```

```python
x
```

```python
y
```

```python
z = tf.add(x,y)
z
```

```python
x + y
```

```python
z = tf.subtract(x,y)
z
```

```python
x - y
```

```python
z = tf.divide(x,y)
z
```

```python
x/y
```

```python
z = tf.multiply(x,y)
z
```

```python
x * y
```

```python
x
```

```python
y
```

```python
z = tf.tensordot(x,y,axes=1)
z
```

```python
tf.reduce_sum(x*y,axis=0)
```

```python
x[:]
```

```python
x[1:]
```

```python
x[1:3]
```

```python
x[::2]
```

```python
x[::-1]
```
