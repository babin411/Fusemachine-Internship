<h3>The .agg() method</h3>

```python
def pct30(column):
    return column.quantile(0.3)
```

```python
dogs['Weight(kg)'].agg(pct30)
```

<h3>Summarizing on multiple columns</h3>

```python
dogs[['Weight(kg)', 'Height(cm)']].agg(pct30)
```

<h3>Multiple Summaries

```python
def pct40(column):
    return column.quantile(0.4)
```

```python
dogs['Weight(kg)'].agg([pct30, pct40])
```

<h3>Cumulative Sum</h3>

```python
dogs['Weight(kg)']
```

```python
dogs['Weight(kg)'].cumsum()
```

<h3>Grouped Summaries</h3>

```python
dogs.groupby('Color')['Weight(kg)'].mean()
```

<h3>Multiple Grouped Summaries</h3>

```python
dogs.groupby('Color')['Weight(kg)'].agg([min, max,sum])
```

```python
dogs.groupby('Color')[['Weight(kg)', 'Height(cm)']].agg([min, max,sum])
```

<h3>Grouping by multple variables</h3>

<h3>Many Groups, Many Summaries</h3>

```python
dogs.groupby(['Color', 'Breed'])[['Weight(kg)', 'Height(cm)']].mean()
```
