Template Matching is the method of searching and finding the location of a template image inside a larger image. 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path = os.path.join(os.getcwd(), 'data')
path
```

```python
img = cv2.imread(os.path.join(path, 'messi5.jpg'))
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
template = cv2.imread(os.path.join(path, 'messi_face.jpg'),0)
```

```python
plt.imshow(img)
```

```python
plt.imshow(template)
```

```python
w,h = template.shape[::-1]
```

```python
print(f'Width: {w}\nHeight: {h}')
```

```python
res = cv2.matchTemplate(grey_img, template,cv2.TM_CCOEFF_NORMED)
print(res)
```

```python
print(len(res))
```

```python
lst= np.arange(1,6)
print(np.where(lst<=3,lst,0))
```

```python
lst= np.arange(1,6)
print(np.where(lst<=3))
```

```python
threshold = 0.9
loc = np.where(res >=threshold)
loc
```

```python
for pt in zip(*loc[::-1]):
    cv2.rectangle(img,pt,(pt[0]+w,pt[1]+h),(0,0,255),2)
```

```python
cv2.imshow('Image',img)
if cv2.waitKey() & 0xFF==ord('q'):
    cv2.destroyAllWindows()
```
