```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# img = cv2.imread('data/opencv-logo-white.png')
img = cv2.imread('lena_copy.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Homogenous filter is the most simple filter, where each pixel is the mean of its kernel neighbors
A kernel, convolution matrix or mask is a small matrix used for blurring, sharpening, embossing, edge detection and many more.

```python
kernel = np.ones((5,5), np.float32) / 25
```

```python
kernel
```

```python
dst = cv2.filter2D(img,-1,kernel)
blur = cv2.blur(img, (5,5))
gaussian = cv2.GaussianBlur(img,(5,5),0)
median = cv2.medianBlur(img, 5)
bilateral = cv2.bilateralFilter(img,9,75,75) #highly effective in noise removal while keeping the edge sharp
```

```python
titles = ['image', '2d Convolution', 'blurr', 'Gaussian Blurr', 'Median Blur',
         'Bilateral']
images = [img, dst, blur, gaussian,median, bilateral]
```

```python
plt.figure(figsize=(13,13))
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
