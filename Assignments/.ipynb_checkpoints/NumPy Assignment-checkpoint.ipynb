### Considering that rows represent each data point and columns represent the features which we will feed to a ML model, perform the following feature scaling operations (without SKLearn):

#### Normalization

Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1.
<br/>
<b>y = (x - min)/(max-min)</b>

```python
new_matrix
```

```python
normalized_matrix= (new_matrix - new_matrix.min()) / (new_matrix.max() - new_matrix.min())
normalized_matrix
```

#### Standardization

Standardizing a dataset involves rescaling the distribution of values so that the mean of observed values is 0 and the standard deviation is 1.<br/>
<b>y=x-mean/standard deviation</b>

```python
new_matrix
```

```python
standardized_matrix = (new_matrix - new_matrix.mean())/new_matrix.std()
standardized_matrix
```
