## Create three arrays: x, y, and z where:
<ul>
    <li> x contains 12 evenly spaced numbers between 1 and 20</li>
    <li> y contains the values of x multiplied by 2 i.e 2*x</li>
    <li> z contains the squares of elements of x i.e x**2</li>
</ul>

```python
x = np.linspace(1,20,12)
x
```

```python
y = x * 2
y
```

```python
z = x ** 2
z
```

#### a. Plot y vs  x, set the label of x-axis as 'x', the label of y-axis as 'y' and the title of the plot as 'y=2x'. Increase the x-axi range to (0,25) and y-axis-range to (0,45).

```python
fig,ax = plt.subplots()
plt.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.xlim(0,26)
plt.ylim(0,450)
plt.title('y = 2x')
```

#### b. Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two functions using legend.

```python
fig,ax = plt.subplots()
plt.plot(x,y,label='2*x',color='red')
plt.plot(x,z, label='x**2',color='blue')
ax.set_xlabel('x')
plt.xlim(0,26)
plt.ylim(0,450)
plt.legend()
```

#### c. Create a figure containing two subplots. Set the size of the figure to be 12 x 6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))
ax[0].plot(x,y)
ax[0].set_title('y=2x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x,z,color='red')
ax[1].set_title('z=x**2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
```

#### d. Redoing all the above plots using seaborn

```python
fig = sns.lineplot(x,y)
fig.set_title('y=2x')
fig.set_xlabel('x')
fig.set_ylabel('y')
fig.set_xlim(0,25)
fig.set_ylim(0,450)
```

```python
fig,ax = plt.subplots()
sns.lineplot(x,y,ax=ax, color='blue',label='y=2x')
ax.set_xlim(0,25)
ax.set_ylim(0,100)
ax2 = ax.twinx()
sns.lineplot(x,z,ax=ax2, color='red',label='z=x**2')
plt.legend()
```
