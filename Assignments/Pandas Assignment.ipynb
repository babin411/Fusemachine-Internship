# 3. Create Two DataFrame (First Dataframe having 50 random rows and second dataframe having 40 random rows). Each dataframe should have 3 columns including a common column (example: id) and 36 rows of data in the second dataframe should be identical to the first dataframe.

```python
data1 = np.random.randint(1,100,(50,3))
data2 = np.random.randint(1,100,(40,3))
```

```python
df1 = pd.DataFrame(data = data1, columns=list('ABC'))
df2 = pd.DataFrame(data = data2, columns = list('CDE'))
```

```python
df1.head(n=2)
```

```python
df2.head(n=2)
```

```python
df2[:36] = df1[:36]
df2
```

```python
df1[:36].values == df2[:36].values
```

### Merge both the DataFarme using:

```python
df1.head(n=5)
```

```python
df2.head(n=5)
```

i. Inner

```python
df1.merge(df2, left_on='C', right_on='C', how='inner',suffixes=('_df1', '_df2'))
```

ii. Outer

```python
df1.merge(df2, left_on='C', right_on='C', how='outer',suffixes=('_df1', '_df2'))
```

iii. Left

```python
len(df1)
```

```python
len(df2)
```

```python
df1.merge(df2, left_on='C', right_on='C', how='left',suffixes=('_df1', '_df2'))
```

iv. Right

```python
df1.head(n=5)
```

```python
df2.head(n=5)
```

```python
df1.merge(df2, left_on='C', right_on='C', how='right',suffixes=('_df1', '_df2'))
```

### Use join() to combine the dataframe

```python
df1.head(n=3)
```

```python
df2.head(n=3)
```

#### Using Right Join

```python
df1.join(df2,how='right', lsuffix='_df1', rsuffix='_df2')
```

#### Using Inner Join

```python
df1.join(df2,how='inner', lsuffix='_df1', rsuffix='_df2')
```

#### Using Outer Join

```python
df1.join(df2,how='outer', lsuffix='_df1', rsuffix='_df2')
```
